# Exercise 1.2 — Yield per die

### Die Yield

Die yield represents the probability that a manufactured chip (die) is functional and free of defects after fabrication. Since defects can occur during wafer production, not every die produced will work correctly.

A higher die yield means more good chips per wafer, reducing manufacturing cost and improving production efficiency.

The formula used in these exercises is:

Die Yield = Wafer Yield × 1 / (1 + Defects per Unit Area × Die Area)^N

Since the book assumes:

Wafer Yield = 100% = 1

we simplify to:

Die Yield = 1 / (1 + Defects per Unit Area × Die Area)^N

Where:

- Defects per Unit Area = defect density of the manufacturing process
- Die Area = physical chip area
- N = complexity factor related to manufacturing sensitivity
- Wafer Yield = percentage of usable wafers (assumed as 100% in these exercises)

### Base Table Used in the Calculations

| Chip | Die Size (mm²) | Estimated Defect Rate (per cm²) | N | Manufacturing Size |
|---|---:|---:|---:|---:|
| GPU A | 800 | 0.011 | 36 | 4 nm |
| GPU B | 2 × 850 | 0.011 | 37 | 4 nm |
| CPU A | 270 | 0.011 | 36 | 4 nm |
| CPU B (Compute) | 12 × 75 | 0.010 | 35 | 5 nm |
| CPU B (I/O) | 1 × 400 | 0.010 | 34 | 6 nm |

### Important Unit Conversion

The defect rate is given in defects per cm², while die size is given in mm².

1 cm² = 100 mm²

Therefore:

Die Area (cm²) = Die Area (mm²) / 100

```python
def calculate_die_yield(
    wafer_yield,
    defects_per_cm2,
    die_area_cm2,
    process_complexity_factor
):
    """
    Calculate die yield using the book formula:

    Die Yield =
        Wafer Yield × 1 / (1 + Defects per Unit Area × Die Area)^N

    Where:

    - wafer_yield = physical dies available from the wafer
    - defects_per_cm2 = defect density of the process
    - die_area_cm2 = chip area in cm²
    - process_complexity_factor = manufacturing sensitivity factor (N)
    """

    defect_impact = (
        1 + (defects_per_cm2 * die_area_cm2)
    ) ** process_complexity_factor

    die_yield = wafer_yield / defect_impact

    return die_yield
```

# Exercise 1.2 — Chip Fabrication Cost

### Dies per Wafer

Before calculating yield, we first estimate how many dies (chips) can physically fit on a wafer.

Because chips near the wafer edge are partially cut and unusable, we use the corrected formula:

Dies per Wafer = (π × (wafer diameter / 2)² / Die Area) − (π × wafer diameter / √(2 × Die Area))

Where:

- wafer diameter = wafer size (300 mm in these exercises)
- Die Area = physical chip area

For these exercises:

- Wafer diameter = 300 mm = 30 cm

The first term estimates the ideal number of dies using total wafer area.

The second term corrects for dies lost at the wafer edge.

---

### Die Yield

Die yield represents the number of functional chips (dies) produced per wafer after considering manufacturing defects.

Since defects occur during fabrication, not every die produced will work correctly.

A higher die yield means more good chips per wafer, reducing manufacturing cost and improving production efficiency.

The formula used in these exercises is:

Die Yield = Wafer Yield × 1 / (1 + Defects per Unit Area × Die Area)^N

Where:

- Defects per Unit Area = defect density of the manufacturing process
- Die Area = physical chip area
- N = complexity factor related to manufacturing sensitivity
- Wafer Yield = number of dies physically available from the wafer

This means:

1. First calculate Dies per Wafer
2. Then apply the defect formula to obtain the number of good chips

---

### Important Unit Conversion

The defect rate is given in defects per cm², while die size is given in mm².

1 cm² = 100 mm²

Therefore:

Die Area (cm²) = Die Area (mm²) / 100

Example:

- GPU A = 800 mm² = 8 cm²
- CPU A = 270 mm² = 2.7 cm²

```python
# =========================
# 1.2 (a)
# =========================
import math


def calculate_max_dies_per_wafer(wafer_diameter_cm, die_area_cm2):
    """
    It calculates the maximum number of dies (chips) that can physically fit on a wafer.

    Formula:
    Dies per Wafer =
        (π * (wafer_diameter / 2)^2) / die_area
        -
        (π * wafer_diameter) / sqrt(2 * die_area)
    """

    wafer_total_area = math.pi * ((wafer_diameter_cm / 2) ** 2)
    ideal_number_of_dies = wafer_total_area / die_area_cm2
    edge_loss_correction = (math.pi * wafer_diameter_cm) / math.sqrt(2 * die_area_cm2)

    max_dies_per_wafer = ideal_number_of_dies - edge_loss_correction

    return max_dies_per_wafer


wafer_diameter_cm = 30          # 300 mm = 30 cm
gpu_a_die_area_cm2 = 8          # 800 mm² = 8 cm²
gpu_a_profit_per_good_chip = 1000

gpu_a_max_dies_per_wafer = calculate_max_dies_per_wafer(
    wafer_diameter_cm,
    gpu_a_die_area_cm2
)

print(gpu_a_max_dies_per_wafer)


gpu_a_good_dies_per_wafer = gpu_a_max_dies_per_wafer * gpu_a_die_yield_per_good_wafer

profit_per_wafer_gpu_a = math.floor(
    gpu_a_good_dies_per_wafer * gpu_a_profit_per_good_chip
)

print(profit_per_wafer_gpu_a)

# Answer:
# The profit per wafer of GPU A chips is about ~3K
```

## Exercise 1.3 — Yield, Binning, and Product Segmentation

Modern processors are composed of multiple cores integrated into a single die. Due to manufacturing defects, not all cores may function correctly.

Instead of discarding partially defective chips, manufacturers use **binning** to create lower-tier products with fewer active cores.

In this exercise, we analyze how defect rates impact the distribution of usable CPUs with different numbers of functional cores.

---

### Assumptions

- CPU A contains **8 identical cores**
- Each core occupies approximately **1/8 of the total die area**
- Defects are randomly distributed across the die
- A core is considered functional only if it is completely defect-free

The yield of a region is modeled using:

Y = e^(-D × A)

Where:

- D = defect density (defects per unit area)  
- A = area of the region  

---

### Yield Scaling

Since each core occupies a fraction of the total die area, the yield of a single core can be approximated as:

Y_core ≈ (Y_chip)^(A_core / A_total)

For this exercise:

Y_core ≈ (Y_chip)^(1/8)

---

#### Exercise 1.3(a)

Given the die yield of CPU A:

- Y_chip ≈ 0.348

Estimate the probability (yield) of obtaining CPUs with:

- 8 functional cores (A8)  
- 4 to 7 functional cores (A4 category)  
- 2 to 3 functional cores (A2 category)  
- exactly 1 functional core (A1 category)  

Assume cores fail independently and use a binomial model.

---

#### Exercise 1.3(b)

From an economic and profitability perspective, discuss:

- Which CPU categories (A8, A4, A2, A1) should be commercialized  
- Which should be discarded  

Consider:

- manufacturing yield distribution  
- expected performance of each category  
- packaging and testing costs  

```python
import math
from fractions import Fraction

# =========================
# 1.3 (a)
# =========================

# Given:
# Y_chip ≈ 0.348

# Yield of one core (1/8 of area)
yield_of_one_eighth_of_the_area = cpu_a_die_yield_per_good_wafer ** Fraction(1, 8)
print(f'Single core yield: {yield_of_one_eighth_of_the_area:.6f}')

# -------------------------
# A8 (8 cores working)
cpu_a8_yield = cpu_a_die_yield_per_good_wafer
print(f'CPU A8 yield: {cpu_a8_yield:.6f}')

# -------------------------
# A4 (4 to 7 cores working)
cpu_a4_yield = sum(
    math.comb(8, k) * (yield_of_one_eighth_of_the_area ** k) * ((1 - yield_of_one_eighth_of_the_area) ** (8 - k))
    for k in [4, 5, 6, 7]
)
print(f'CPU A4 yield: {cpu_a4_yield:.6f}')

# -------------------------
# A2 (2 or 3 cores working)
cpu_a2_yield = sum(
    math.comb(8, k) * (yield_of_one_eighth_of_the_area ** k) * ((1 - yield_of_one_eighth_of_the_area) ** (8 - k))
    for k in [2, 3]
)
print(f'CPU A2 yield: {cpu_a2_yield:.6f}')

# -------------------------
# A1 (1 core working)
cpu_a1_yield = math.comb(8, 1) * (yield_of_one_eighth_of_the_area ** 1) * ((1 - yield_of_one_eighth_of_the_area) ** 7)
print(f'CPU A1 yield: {cpu_a1_yield:.10f}')


# =========================
# Results summary
# =========================
# Single core yield ≈ 0.876
# A8 ≈ 0.348
# A4 ≈ 0.653
# A2 ≈ 0.00117
# A1 ≈ 0.00000305


# =========================
# 1.3 (b)
# =========================

# From an economic and profitability perspective, the optimal strategy is to:
#
# - Sell A8 (fully functional CPUs), as they provide the highest value.
# - Sell A4 CPUs (4–7 working cores), since they represent a large portion of
#   the yield and still offer good performance, maximizing wafer utilization.
#
# Lower-tier CPUs:
# - A2 has extremely low probability (~0.1%) and limited performance.
# - A1 is practically negligible (~3e-06) and not economically viable.
#
# Due to packaging, testing, and distribution costs, it does not make sense
# to commercialize A1 (and likely A2). These chips are typically discarded.
#
# Therefore, focusing on A8 and A4 maximizes profitability while minimizing waste.
```

## Exercise 1.4 — Yield Improvement with Redundancy in Chiplet-Based GPUs

Modern GPUs are composed of multiple **Streaming Multiprocessors (SMs)** integrated into a chiplet. Due to manufacturing defects, not all SMs may function correctly.

Instead of requiring all SMs to be defect-free, manufacturers can introduce **redundancy**, allowing a limited number of defective SMs while still considering the chiplet functional.

- GPU B chiplet contains **85 identical SMs**
- Each SM occupies approximately **1/85 of the total chiplet area**
- Defects are randomly distributed across the chiplet
- An SM is functional only if it is defect-free
- A chiplet is considered functional if it has **at least 80 working SMs**
- Each final GPU product requires **2 chiplets**

The yield of a region is modeled using:

Y = e^(-D × A)

Where:

- D = defect density (defects per unit area)  
- A = area of the region  

---

### Yield Scaling

Since each SM occupies a fraction of the total chiplet area, the yield of a single SM can be approximated as:

Y_SM ≈ (Y_chiplet)^(A_SM / A_total)

For this exercise:

Y_SM ≈ (Y_chiplet)^(1/85)
