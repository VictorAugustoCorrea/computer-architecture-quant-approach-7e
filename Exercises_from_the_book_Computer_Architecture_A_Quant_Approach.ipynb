# Exercise 1.2 — Yield per die

### Die Yield

Die yield represents the probability that a manufactured chip (die) is functional and free of defects after fabrication. Since defects can occur during wafer production, not every die produced will work correctly.

A higher die yield means more good chips per wafer, reducing manufacturing cost and improving production efficiency.

The formula used in these exercises is:

Die Yield = Wafer Yield × 1 / (1 + Defects per Unit Area × Die Area)^N

Since the book assumes:

Wafer Yield = 100% = 1

we simplify to:

Die Yield = 1 / (1 + Defects per Unit Area × Die Area)^N

Where:

- Defects per Unit Area = defect density of the manufacturing process
- Die Area = physical chip area
- N = complexity factor related to manufacturing sensitivity
- Wafer Yield = percentage of usable wafers (assumed as 100% in these exercises)

### Base Table Used in the Calculations

| Chip | Die Size (mm²) | Estimated Defect Rate (per cm²) | N | Manufacturing Size |
|---|---:|---:|---:|---:|
| GPU A | 800 | 0.011 | 36 | 4 nm |
| GPU B | 2 × 850 | 0.011 | 37 | 4 nm |
| CPU A | 270 | 0.011 | 36 | 4 nm |
| CPU B (Compute) | 12 × 75 | 0.010 | 35 | 5 nm |
| CPU B (I/O) | 1 × 400 | 0.010 | 34 | 6 nm |

### Important Unit Conversion

The defect rate is given in defects per cm², while die size is given in mm².

1 cm² = 100 mm²

Therefore:

Die Area (cm²) = Die Area (mm²) / 100

```python
def calculate_die_yield(
    wafer_yield,
    defects_per_cm2,
    die_area_cm2,
    process_complexity_factor
):
    """
    Calculate die yield using the book formula:

    Die Yield =
        Wafer Yield × 1 / (1 + Defects per Unit Area × Die Area)^N

    Where:

    - wafer_yield = physical dies available from the wafer
    - defects_per_cm2 = defect density of the process
    - die_area_cm2 = chip area in cm²
    - process_complexity_factor = manufacturing sensitivity factor (N)
    """

    defect_impact = (
        1 + (defects_per_cm2 * die_area_cm2)
    ) ** process_complexity_factor

    die_yield = wafer_yield / defect_impact

    return die_yield
```

### Dies per Wafer

Before calculating yield, we first estimate how many dies (chips) can physically fit on a wafer.

Because chips near the wafer edge are partially cut and unusable, we use the corrected formula:

Dies per Wafer = (π × (wafer diameter / 2)² / Die Area) − (π × wafer diameter / √(2 × Die Area))

Where:

- wafer diameter = wafer size (300 mm in these exercises)
- Die Area = physical chip area

For these exercises:

- Wafer diameter = 300 mm = 30 cm

The first term estimates the ideal number of dies using total wafer area.

The second term corrects for dies lost at the wafer edge.

### Die Yield

Die yield represents the number of functional chips (dies) produced per wafer after considering manufacturing defects.

Since defects occur during fabrication, not every die produced will work correctly.

A higher die yield means more good chips per wafer, reducing manufacturing cost and improving production efficiency.

The formula used in these exercises is:

Die Yield = Wafer Yield × 1 / (1 + Defects per Unit Area × Die Area)^N

Where:

- Defects per Unit Area = defect density of the manufacturing process
- Die Area = physical chip area
- N = complexity factor related to manufacturing sensitivity
- Wafer Yield = number of dies physically available from the wafer

This means:

1. First calculate Dies per Wafer
2. Then apply the defect formula to obtain the number of good chips

```python
# 1.2 (a)

import math


def calculate_max_dies_per_wafer(wafer_diameter_cm, die_area_cm2):
    """
    Calcula o número máximo de dies (chips) que cabem fisicamente em um wafer.

    Fórmula:
    Dies per Wafer =
        (π * (wafer_diameter / 2)^2) / die_area
        -
        (π * wafer_diameter) / sqrt(2 * die_area)
    """

    wafer_total_area = math.pi * ((wafer_diameter_cm / 2) ** 2)
    ideal_number_of_dies = wafer_total_area / die_area_cm2
    edge_loss_correction = (math.pi * wafer_diameter_cm) / math.sqrt(2 * die_area_cm2)

    max_dies_per_wafer = ideal_number_of_dies - edge_loss_correction

    return max_dies_per_wafer


wafer_diameter_cm = 30          # 300 mm = 30 cm
gpu_a_die_area_cm2 = 8          # 800 mm² = 8 cm²
gpu_a_profit_per_good_chip = 1000

gpu_a_max_dies_per_wafer = calculate_max_dies_per_wafer(
    wafer_diameter_cm,
    gpu_a_die_area_cm2
)

print(gpu_a_max_dies_per_wafer)


gpu_a_good_dies_per_wafer = gpu_a_max_dies_per_wafer * GPU_A_chip_yield

profit_per_wafer_gpu_a = math.floor(
    gpu_a_good_dies_per_wafer * gpu_a_profit_per_good_chip
)

print(profit_per_wafer_gpu_a)

# Answer:
# The profit per wafer of GPU A chips is about ~3K
```

```python
# 1.2 (b)

cpu_a_die_area_cm2 = 2.7        # 270 mm² = 2.7 cm²
cpu_a_profit_per_good_chip = 100

cpu_a_max_dies_per_wafer = calculate_max_dies_per_wafer(
    wafer_diameter_cm,
    cpu_a_die_area_cm2
)

print(cpu_a_max_dies_per_wafer)


cpu_a_good_dies_per_wafer = cpu_a_max_dies_per_wafer * CPU_A_yield_chip

profit_per_wafer_cpu_a = math.floor(
    cpu_a_good_dies_per_wafer * cpu_a_profit_per_good_chip
)

print(profit_per_wafer_cpu_a)

# Answer:
# The profit per wafer of CPU A chips is about ~7.7K
```

```python
# 1.2 (c)

gpu_a_monthly_demand = 5000
cpu_a_monthly_demand = 3000


gpu_a_good_chips_per_wafer = (
    gpu_a_max_dies_per_wafer * GPU_A_chip_yield
)

cpu_a_good_chips_per_wafer = (
    cpu_a_max_dies_per_wafer * CPU_A_yield_chip
)


gpu_a_wafers_needed = (
    gpu_a_monthly_demand / gpu_a_good_chips_per_wafer
)

cpu_a_wafers_needed = (
    cpu_a_monthly_demand / cpu_a_good_chips_per_wafer
)


total_wafers_needed = (
    gpu_a_wafers_needed + cpu_a_wafers_needed
)


print(gpu_a_wafers_needed)
print(cpu_a_wafers_needed)
print(total_wafers_needed)

# Answer:
# To supply the monthly demand, we need about:
# ~1607 wafers for GPU A
# ~38 wafers for CPU A
# Total: ~1646 wafers
#
# Therefore, it is NOT possible to meet demand
# with a factory capacity of only 90 wafers/month.
```

```python
# 1.2 (e)

cpu_b_profit_per_product = 75

compute_chiplets_needed_per_product = 12
io_chiplets_needed_per_product = 1

total_available_wafers = 30


# --------------------------------------------------
# Good chips per wafer (already calculated before)
# --------------------------------------------------

good_compute_chiplets_per_wafer = 865
good_io_chiplets_per_wafer = 143

print(good_compute_chiplets_per_wafer)
print(good_io_chiplets_per_wafer)


# --------------------------------------------------
# Step 1: How many complete CPU B products can be
# made from one wafer of compute chiplets
# --------------------------------------------------

products_per_compute_wafer = (
    good_compute_chiplets_per_wafer
    / compute_chiplets_needed_per_product
)

print(products_per_compute_wafer)

# ≈ 72 products per compute wafer


# --------------------------------------------------
# Step 2: How many compute wafers are needed to
# match one I/O wafer
# --------------------------------------------------

compute_wafers_needed_per_io_wafer = (
    good_io_chiplets_per_wafer
    / products_per_compute_wafer
)

print(compute_wafers_needed_per_io_wafer)

# ≈ 2 compute wafers for each I/O wafer


# --------------------------------------------------
# Step 3: Best allocation of 30 wafers
# --------------------------------------------------

io_wafers_allocated = 10
compute_wafers_allocated = 20

print(io_wafers_allocated)
print(compute_wafers_allocated)


# --------------------------------------------------
# Step 4: Total final products produced
# --------------------------------------------------

total_products_from_compute = (
    compute_wafers_allocated
    * products_per_compute_wafer
)

total_products_from_io = (
    io_wafers_allocated
    * good_io_chiplets_per_wafer
)

total_final_products = math.floor(
    min(total_products_from_compute, total_products_from_io)
)

print(total_final_products)

# ≈ 1430 products


# --------------------------------------------------
# Step 5: Maximum total profit
# --------------------------------------------------

maximum_total_profit = (
    total_final_products
    * cpu_b_profit_per_product
)

print(maximum_total_profit)

# Answer:
# Products per compute wafer ≈ 72
# Chips for I/O per wafer = 143
#
# Best distribution of 30 wafers:
# 20 wafers for compute chiplets
# 10 wafers for I/O chiplets
#
# Total final products ≈ 1430
# Maximum profit ≈ $107,250
```
